```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Defining the function
def f(x1,x2):
    return (x2-x1)**4 + 8*x1*x2 -x1 + x2 + 3
```

```python
x1 = np.linspace(-1,1,60,endpoint=True).reshape((60,1)) # Initialize x1 [-1 ... 1]

# Calculating f values for x1 = -1 and 1 
y11 = [f(i,-1) for i in x1]
y12 = [f(i,1) for i in x1]

#Generating line plot
plt.plot(x1,y11,label="x2 = -1")
plt.plot(x1,y12,label="x2 = 1")
plt.xlabel("x1")
plt.ylabel("Test function")
plt.title("Test function vs x1 for a fixed x2")
plt.legend()
plt.show()
```

```python
#Generating Surface plot
fig = plt.figure(figsize=(10,10))#Initialize figure size
ax = plt.axes(projection='3d')
x1 = np.outer(np.linspace(-1, 1, 60), np.ones(60))
x2 = x1.copy().T # creating x2
ax.plot_surface(x1, x2, f(x1,x2),cmap='viridis', edgecolor='none') # generating result with all possible combinations of x1 and x2
ax.set_title('Surface plot of test function')
ax.set_xlabel("x1")
ax.set_ylabel("x2")
ax.set_zlabel("Test function")
plt.show()
```

```python
#Generating contour plot
X, Y = np.meshgrid(x1,x2)
Z = f(X,Y)
fig, ax = plt.subplots(1, 1)
# plots contour lines
CS = ax.contour(X, Y, Z)
ax.set_title('Contour Plot for test function')
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.clabel(CS)
plt.show()
```
